```python
import torch

list = [[1,2,3],
       [4,5,6]]

tns = torch.Tensor(list)
print(tns.dtype)
print(tns)

tns = torch.FloatTensor(list)
print(tns.dtype)
print(tns)

tns = torch.DoubleTensor(list)
print(tns.dtype)
print(tns)

tns = torch.LongTensor(list)
print(tns.dtype)
print(tns)
```

```python
import numpy as np

arr = np.random.rand(3,4)
print(arr)
print(arr.dtype)

tns = torch.from_numpy(arr)
print(tns)
print(tns.dtype)
```

```python
tns0 = torch.zeros(2,3)
tns1 = torch.ones(3,4)
tnsr = torch.randn(4,5)

print(tns0)
print(tns1)
print(tnsr)
```

```python
print(tnsr)
print("")
tnsr[0,2]=-10
print(tnsr)
print("")
print(tnsr[0:2])
print("")
print(tnsr[:,2])
print("")
print(tnsr[0,2])
```

```python
tnsr1 = torch.randn(2,3)
tnsr2 = torch.randn(2,3)

print(tnsr1)
print(tnsr2)
print("")
print(tnsr1+tnsr2)
print("")
print(tnsr1-tnsr2)
print("")
print(tnsr1*tnsr2)
print("")
print(tnsr1/tnsr2)
print("")
print(torch.mm(tnsr1,tnsr2.T))
```

```python
tnsr = torch.randn(2,3,4)
print(tnsr)
print("")
print(tnsr.size())
print("")
print(tnsr.view(3,8))
print("")
print(tnsr.view(-1))
print("")
print(tnsr.view(tnsr.size(0),-1))
```
